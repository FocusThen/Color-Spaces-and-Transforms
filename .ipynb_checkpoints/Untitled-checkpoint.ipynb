```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image = cv2.imread('ballons.jpg')
print("İmage shape: ", image.shape)
```

```python
plt.imshow(image)
```

```python
image_copy = np.copy(image)

image_copy = cv2.cvtColor(image_copy, cv2.COLOR_BGR2RGB)

plt.imshow(image_copy)
```

```python
r = image_copy[:,:,0]
g = image_copy[:,:,1]
b = image_copy[:,:,2]
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20, 10))

ax1.set_title('Red')
ax1.imshow(r, cmap="gray")

ax2.set_title('Green')
ax2.imshow(g, cmap="gray")

ax3.set_title('Blue')
ax3.imshow(b, cmap="gray")
```

```python
hsv = cv2.cvtColor(image_copy, cv2.COLOR_RGB2HSV)
```

```python
h = hsv[:, :, 0]
s = hsv[:, :, 1]
v = hsv[:, :, 2]
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1,3, figsize=(20, 10))

ax1.set_title('Hue')
ax1.imshow(h, cmap="gray")

ax2.set_title('Saturation')
ax2.imshow(s, cmap="gray")

ax3.set_title('Value')
ax3.imshow(v, cmap="gray")
```

```python
#RGB value
lower_pink = np.array([180, 0, 100])
upper_pink = np.array([255, 255, 230])
```

```python
#HSV value
lower_hue = np.array([160, 0, 0])
upper_hue = np.array([180, 255, 255])
```

```python
#Masks
#RGB MASK
mask_rgb = cv2.inRange(image_copy, lower_pink, upper_pink)

masked_img = np.copy(image_copy)
masked_img[mask_rgb == 0] = [0,0,0]
```

```python
plt.imshow(masked_img)
plt.title('RGB pink')
```

```python
#Masks
#HSV MASK
mask_hsv = cv2.inRange(hsv, lower_hue, upper_hue)

masked_image = np.copy(image_copy)
masked_image[mask_hsv == 0] = [0, 0, 0]
```

```python
plt.imshow(masked_image)
plt.title('HSV pink')
```
